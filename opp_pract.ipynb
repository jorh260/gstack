# **Step 9**

**super()**:is similar to inherentance. it  enable to use of the parent class attributes/actions/method on the child/children classes, this is to prevent repetition, errors, and easy modifications.


this simply means that, to look through the children classes, what attributes/actions/method to they have in common? instead of writing them one by one, and if any modification is needed, it would need to be apply manually.


To prevent that, the "super()" is used. group the common traits into a parent class, and use "super()" to apply it to the the children classes.


## **super() is used, to slice through the trait from parent class to use in the children classes**



**another uses ofthe 'super()' is to extend the functionality of the parent class and child class. eg: parent class as  " describe method", and the childre classes also have "describe method", 'the super() ' is use to ensure that each describe methods of both the parent and child classes are executed.**

````python
# super ()


# creating aparent class to encode what the children clases will have in common
class shape:
  def __init__(self,color,is_filled):
    self.color = color
    self.is_filled = is_filled
  def describe(self):
    print(f'the shape color is {self.color} and {'filled' if self.is_filled else 'not filled'} ')


# creating a child class
class circle (shape): # the parent class is inhereted by the child(circle) class
  def __init__(self,color,is_filled,radius):
   # self.color =color
   # self.is_filled = is_filled (instead of writing this in all the children class, super() is used, to slice through the trait from parent class to use in the children classes)
    super().__init__(color,is_filled) #super() function is used here  to slice out the needed trait from the parent class to the children classes.
    self.radius = radius


class square(shape):
 def __init__(self,color,is_filled,width):

    super().__init__(color,is_filled)
    self.width = width


class triangle(shape):
  def __init__(self,color,is_filled,width,height):

    super().__init__(color,is_filled)
    self.width = width
    self.height= height

# creating object for circle

circle = circle('yellow', True, 4)

# output
print(circle.color)
print(circle.is_filled)
print(circle.radius)
````

````python

# dunder method

class book :
  def __init__(self, title, author,num_pages):

     self.title = title
     self.author = author
     self.num_pages = num_pages
   #the dunder method, in this case, its use to modify / customise cla
  def __str__(self):
    return f'{self.title} by {self.author} and it as {self.num_pages} pages'
  # dunder method for equal, which is __eq__
  def __eq__(self, other):
    return self.title == other.title and self.author == other.author and self.num_pages == other.num_pages

  # another dunder method example: less than as __lt__()
  def __lt__(self,other):
    return self.num_pages < other.num_pages

  # another dunder method example: addition as __add__()
  def __add__(self,other):
    return self.num_pages + other.num_pages

  # another dunder method example: contains as __contains__()
  def __contains__(self,keyword):
    return  keyword in self.title or keyword in self.author


   # another dunder method example: getitem  as    __getitem__()
  def __getitem__(self,key): # Corrected: Changed __getitem___ to __getitem__
    if key == 'title':
       return self.title
    elif key =='author' :
      return self.author
    elif key=='num_pages':
      return self.num_pages
    else:
     return f'key {key} was mot found'


#objects
book1 = book('sandy dies','andrew lewy', 150 )
book2 = book('the herry potter ',' ben slayer', 320 )
book3 = book('dance in the rain ',' bintiw  tutu', 200 )


print(book1)   # output: sandy dies by andrew lewy and it as 150 pages
# another dunder method: to check if two object are equal.
print(book1 == book2) # output: False
print(book1 < book2)
print(book1 + book2)
print('lion' in book3)
print(book1['title'])
````

# **step 18**



Property Decorator:

Makes a method behave like an attribute.

Structure:


"@property

def method(self):  "

Memory Trick:

Method acting like variable:  **obj.method**


Most Important Difference
Normal Method	Property  **obj.method()**







````python
#property decorator

class rectangle :
  def __init__(self, width, height): # Corrected __iniy__ to __init__
    self._width = width # a slight change is done to the 'attribute' which is the addition of underscore, to make it private
    self._height =height # same thing applies here too


# this is to turn the "width" and "height" to be use as variable,eg:  this "rectangle.width" instead of "rectangle.width()"
  @property
  def width(self):
    return f'{self._width:.1f}cm'

  @property
  def height(self):
    return f'{self._height:.1f}cm'

# setter method : it enable addition or append of attribute values based on a condition designs
  @width.setter
  def width(self,new_width):
    if new_width > 0:
       self._width = new_width
    else:
        return f'width most be greater than zero'

  @height.setter
  def height(self,new_height): # Corrected: 'def width' changed to 'def height'
    if new_height > 0:
       self._height = new_height
    else:
        return f'height  most be greater than zero'

#delete an attribute:
  @width.deleter
  def width(self):
    del self._width
    print('width has been deleted')

  @height.deleter
  def height(self):
    del self._height
    print('height has been deleted')


#object

rec = rectangle(3,4)

rec.width = 8 # Corrected: 'rectangle.width' changed to 'rec.width'
rec.height = 7 # Corrected: 'rectangle.height' changed to 'rec.height'

print(rec.height)
print(rec.width)
del rec.width
del rec.height
````

````python



class balance_error(Exception):
    """Base class for other exceptions"""
    pass


class BankAccount:
    def __init__(self,initialbalance,accname):
        self.balance = initialbalance
        self.name = accname
        #account created print
        print(f"\nAccount '{self.name}'  is created   \nBalance = ${self.balance} ")


    def getbalance(self):
        return f"\nAccount'{self.name}' your account balance = ${self.balance:.2f}"

    def deposit(self,amount):
        self.balance += amount
        print(self.getbalance())
        print('deposit successful')

    def balance_check(self,amount):
        if self.balance >= amount:
           return
        else:
            raise balance_error("Insufficient funds")


    def withdraw(self,amount):
       try:
            self.balance_check(amount)
            self.balance -= amount
            print(self.getbalance())
            print('withdrawal successful')

        except balance_error as error:
           print(f'withdrawal failed: {error}')







````
